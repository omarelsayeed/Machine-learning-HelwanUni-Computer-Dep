# Introduction 
with reinforcement learning we can train AIs to perform complicated tasks, but unlike other techniques, we only have to tell them at the very end of the task if they succeeded, and then ask them to tell us how well they did. for instance, if we want an AI to learn to walk, we give them a reward if they're both standing up and moving forward  and then, figure out what steps they took to get to that point, the longer the AI stands up and moves forward, the longer it's walking and the more reward it gets, so the key to reinforcement learning is just **trial and error**. the reward for an AI system is just a small positive signal that basically tells it **good job!** and **Do that again**.

# Markoov decision process 


in MDP, a loyal subject that is going to learn named **AGENT** makes predictions or performs **ACTIONS**, like moving a tiny bit forward or even picking the next best move in a game. it performs **ACTIONS** based on its current inputs, which we call it **STATE** and then it takes a **reward** after executing that action that changed the state from **s1** to **s2**.


so we can say that: at time step t=0 (the first time step), the environment (including the agent) is in some state st=s0, takes an action at=a0 and receives and reward rt=r0 and the environment (including the agent) moves to a next state st+1=s0+1=s1, which will also be the state that the environment will be in at the next time step, t+1, hence the notation st+1. 

**note:**

unlike the supervised learning, we don't have a previous knowledge(i mean labeled data) about the right action  untill it's completely done by the agent.
 



# Values and policies
 
#### we need **values and policies** to get anything done in reinforcement learning, let's add a simple defintion for both terminologies:

### policy:

the probability that an agent will select a specific action from a specific state.

a policy is a function that maps a given state to probabilities of selecting each possible action from that state. we use the symbol **pi** to refer to a policy.


the above symbol means that : the policy at time t is the probability of selecting action **t** at state **S**


### value function

tells "how good" a specfic action or a specific state. in order to maximize the expected retrun, the value function tells us whether a specific action selected over another in a given state may increase or decrease the agent's return, thus, the agent can maximize the expected return. 


Value function is a function of **states** or a **state-action** pairs that estimates how good it is for an agent to be in a given state or to perform an action in a given state, and this is given in terms of **expected return**.

as long as the value function wants to maximize the **expected return**, it must be a fuction of **specific ways of acting**(actions at specific states) and influenced by the **policy** it's following. 


### Types of value functions:
